## Exploratory Data Analysis


### Total Cases all over the world

```python
df_world_over_time = df_world.groupby(["date"])[["total_cases","new_cases","total_deaths","new_deaths"]].sum().reset_index().sort_values("date",ascending=True).reset_index(drop=True)
df_world_over_time
```

```python
# change date from object to datetime64
df_world_over_time ['date'] = pd.to_datetime(df_world_over_time ['date'])
df_world_over_time.info()
```

```python
# Total cases of Covid-19 cases over time in the word
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])
plt.title('Total cases of Covid-19 cases over time in the word', fontsize=16)
plt.xlabel('Months', fontsize=16)
plt.ylabel('Total cases', fontsize=16)
plt.xticks(rotation=90)
```

```python
# Total Deaths of Covid-19 cases over time in the word
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_deaths'])
plt.title(' Total Deaths of Covid-19 Deaths cases over time in the world', fontsize=16)
plt.xlabel('Months', fontsize=16)
plt.ylabel('Total Deaths', fontsize=16)
plt.xticks(rotation=90)
```

```python
# New Cases of Covid-19 cases over time in the word
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['new_cases'],marker='^',c='r',ls='--', lw=.25, markersize=5 )
plt.title('New Cases of Covid-19 New Cases over time in the world', fontsize=16)
plt.xlabel('Months', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
plt.xticks(rotation=90)
```

```python
# New deaths of Covid-19 Deaths cases over time in the world
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['new_deaths'],marker='o',c='b',ls='--', lw=.25, markersize=5)
plt.title(' New deaths of Covid-19 Deaths cases over time in the world', fontsize=16)
plt.xlabel('Months', fontsize=16)
plt.ylabel('New deaths', fontsize=16)
plt.xticks(rotation=90)
```

```python
# Comparing between (Total Cases,Total Deaths, New Cases, New Deaths)
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'], label='Total Cases')
plt.plot(df_world_over_time['date'], df_world_over_time['total_deaths'], label='Total Deaths')
plt.plot(df_world_over_time['date'], df_world_over_time['new_cases'], label='New cases')
plt.plot(df_world_over_time['date'], df_world_over_time['new_deaths'], label='New Deaths')

plt.title('Comparing between Total Cases,Total Deaths, New Cases, New Deaths', fontsize=16)
plt.xlabel('Months', fontsize=16)
plt.legend(loc=0)
plt.show()
```

### New cases of January 2022 in Egypt

```python
df_egypt_mod = df_egypt[['date', 'new_cases']]
df_egypt_mod
```

```python
# Summery about Data 
df_egypt_mod.info()
```

```python
# change date from object to datetime64
pd.to_datetime(df_egypt_mod['date'])
```

```python
# Summery about Data after change date from object to datetime64
df_egypt_mod.info()
```

```python
# make date index
df_egypt_mod.set_index('date')
```

```python
# New cases in January 2021 in Egypt by days
df_jan_egypt_21 = df_egypt_mod[(df_egypt_mod.date >= '2021-01-01') & (df_egypt_mod.date <= '2021-01-31')].reset_index(drop=True)
df_jan_egypt_21
```

```python
# Total New cases in January 2021 in Egypt
df_jan_egypt_21['new_cases'].sum()
```

```python
# New cases in January 2021 in Egypt by days
plt.figure(figsize=(20,10))
plt.plot(df_jan_egypt_21['date'], df_jan_egypt_21['new_cases'])

plt.title('New Cases of Covid-19 New Cases in Egypt January 2021 ', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
plt.xticks(rotation=90)
```

```python
# New cases in January 2022 in Egypt by days
df_jan_egypt_22 = df_egypt_mod[(df_egypt_mod.date >= '2022-01-01') & (df_egypt_mod.date <= '2022-01-31')].reset_index(drop=True)
df_jan_egypt_22
```

```python
# Total New cases in January 2022 in Egypt
df_jan_egypt_22['new_cases'].sum()
```

```python
# New cases in January 2022 in Egypt by days

plt.figure(figsize=(20,10))
plt.plot(df_jan_egypt_22['date'], df_jan_egypt_22['new_cases'])

plt.title('New Cases of Covid-19 New Cases in Egypt January 2022 ', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
plt.xticks(rotation=90)
```

```python
# comparing between New cases in January 2021 and 2022
plt.figure(figsize=(20,10))
plt.bar(df_jan_egypt_21['date'], df_jan_egypt_21['new_cases'], label='New Cases Jan 2021')
plt.bar(df_jan_egypt_22['date'], df_jan_egypt_22['new_cases'], label='New Cases Jan 2022')

plt.title('Comparing between New cases in January 2021 and 2022 ', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
plt.legend(loc=0)
plt.show()
plt.xticks(rotation=90)
```
